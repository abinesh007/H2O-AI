Location Fetching by IP

```python
import requests

def get_location_from_ip():
    """
    Fetch approximate location (latitude, longitude, city, country) using the public IP address.
    Returns a dictionary with lat, lon, city, country, or None if failed.
    """
    try:
        # ip-api.com returns JSON with location data. No API key needed.
        response = requests.get("http://ip-api.com/json/", timeout=10)
        response.raise_for_status()
        data = response.json()

        if data.get("status") == "success":
            location = {
                "lat": data["lat"],
                "lon": data["lon"],
                "city": data["city"],
                "country": data["country"],
                "ip": data["query"]
            }
            return location
        else:
            print("❌ Location lookup failed:", data.get("message", "Unknown error"))
            return None
    except requests.exceptions.RequestException as e:
        print(f"❌ Network error: {e}")
        return None

def main():
    print("🔍 Detecting your location by IP address...")
    loc = get_location_from_ip()

    if loc:
        print(f"\n📍 Your approximate location:")
        print(f"   IP address: {loc['ip']}")
        print(f"   Latitude:  {loc['lat']}°")
        print(f"   Longitude: {loc['lon']}°")
        print(f"   City:      {loc['city']}")
        print(f"   Country:   {loc['country']}")
    else:
        print("\n⚠️  Could not determine your location. Check your internet connection.")

if __name__ == "__main__":
    main()
```

Fetch Weather from Lat and Log

```python
import requests

def get_weather_by_coords(lat, lon):
    """
    Fetch current weather data from Open-Meteo API (free, no API key required).
    Returns a dictionary with temperature, humidity, wind speed, pressure, UV index, etc.
    """
    # Parameters for Open-Meteo API
    params = {
        "latitude": lat,
        "longitude": lon,
        "current": "temperature_2m,relative_humidity_2m,wind_speed_10m,pressure_msl,uv_index",
        "timezone": "auto"
    }

    try:
        response = requests.get("https://api.open-meteo.com/v1/forecast", params=params, timeout=10)
        response.raise_for_status()
        data = response.json()

        # Extract the data we need
        current = data.get("current", {})
        weather = {
            "temperature": current.get("temperature_2m"),
            "humidity": current.get("relative_humidity_2m"),
            "wind_speed": current.get("wind_speed_10m"),
            "pressure": current.get("pressure_msl"),
            "uv_index": current.get("uv_index")
        }
        return weather
    except requests.exceptions.RequestException as e:
        print(f"❌ Network error while fetching weather: {e}")
        return None
    except KeyError as e:
        print(f"❌ Unexpected API response format: {e}")
        return None


def main():
    print("\n🌤️ Fetching current weather data...")
    weather = get_weather_by_coords(10.3843,78.8209)

    if not weather:
        print("❌ Could not retrieve weather data.")
        return

    print("\n--- Weather Report ---")
    print(f"🌡️ Temperature:   {weather['temperature']}°C")
    print(f"💧 Humidity:      {weather['humidity']}%")
    print(f"💨 Wind speed:    {weather['wind_speed']} km/h")
    print(f"🌬️ Pressure:      {weather['pressure']} hPa")
    print(f"☀️ UV Index:      {weather['uv_index']} (higher = stronger sun protection needed)")


if __name__ == "__main__":
    main()
```
